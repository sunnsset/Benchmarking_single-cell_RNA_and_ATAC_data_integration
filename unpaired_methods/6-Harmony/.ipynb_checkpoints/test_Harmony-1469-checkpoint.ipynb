```r
library(argparse)
library(Seurat)
library(harmony)
library(yaml)
library(dplyr)
library(harmony)
library(patchwork)
library(tidyverse)
```

```r
results_dir <- "/home/xcx/results/1469/6-Harmony"

counts.rna <- t(as.matrix(read.csv(file = paste0("/home/xcx/MYBenchmark-datas/1469/counts_rna.csv"), header = TRUE, row.names="X")))
# counts.rna <- as.matrix(read.csv(file = paste0("/home/xcx/SNARE-GSE126074/1469/counts_rna.csv"), header = TRUE, row.names="X"))
rownames(counts.rna) <- paste("Gene-", seq(1, dim(counts.rna)[1]), sep = "")
colnames(counts.rna) <- paste("Cell-", seq(1, dim(counts.rna)[2]), sep = "")

counts.atac <- t(as.matrix(read.csv(file = paste0("/home/xcx/MYBenchmark-datas/1469/counts_atac.csv"), header = TRUE, row.names="X")))
# counts.atac <- as.matrix(read.csv(file = paste0("/home/xcx/SNARE-GSE126074/1469/counts_atac.csv"), header = TRUE, row.names="X"))
rownames(counts.atac) <- paste("Region-", seq(1, dim(counts.atac)[1]), sep = "")
colnames(counts.atac) <- paste("Cell-", seq(dim(counts.rna)[2] + 1, dim(counts.rna)[2] + dim(counts.atac)[2]), sep = "")

label.rna <- read.table(file = paste0("/home/xcx/MYBenchmark-datas/1469/anno.txt"), sep = "\t", header = FALSE)
label.atac <- read.table(file = paste0("/home/xcx/MYBenchmark-datas/1469/anno.txt"), sep = "\t", header = FALSE)
```

```r
rna.so <- CreateSeuratObject(counts = counts.rna, assay = "RNA", project = "scRNA-seq")
atac.so <- CreateSeuratObject(counts = counts.atac, assay = "ATAC", project = "scATAC-seq")

combined.so <- merge(rna.so, atac.so)
combined.so
head(combined.so)
table(combined.so[[]]$orig.ident)
```

```r
start_time <- proc.time()
combined.so <- NormalizeData(combined.so)
combined.so <- FindVariableFeatures(combined.so, selection.method = "vst", verbose = FALSE)
combined.so <- ScaleData(combined.so)
combined.so <- RunPCA(combined.so, seed.use = 0, verbose = FALSE)

combined.so
head(combined.so)
```

```r
combined.so_harmony <- RunHarmony(combined.so, group.by.vars = "orig.ident", plot_convergence = TRUE, max.iter.harmony = 50, lambda = 0.1)

elapsed_time <- proc.time() - start_time
```

```r
elapsed_time
```

```r
# clustering
combined.so_harmony <- FindNeighbors(combined.so_harmony, reduction = "harmony", dims = 1:15) %>% FindClusters(resolution = 1)
# umap/tsne
# combined.so_harmony<- RunTSNE(combined.so_harmony, reduction = "harmony", dims = 1:15)
combined.so_harmony <- RunUMAP(combined.so_harmony, reduction = "harmony", dims = 1:15)
```

```r
umap_integrated1 <- DimPlot(combined.so_harmony, reduction = "umap", group.by = "orig.ident")
umap_integrated2 <- DimPlot(combined.so_harmony, reduction = "umap", label = TRUE)

umap_tsne_integrated <- CombinePlots(list(umap_integrated1,umap_integrated2),ncol=2)
umap_tsne_integrated
ggsave("/home/xcx/results/1469/6-Harmony/umap_tsne_integrated.pdf",umap_tsne_integrated,wi=25,he=15)
```

```r
harmony_embeddings <- Embeddings(combined.so_harmony, reduction='harmony')
```

```r
library(data.table)
# data_to_write_out <- as.data.frame(as.matrix(coembed[['RNA']]))
data_to_write_out <- as.data.frame(as.matrix(harmony_embeddings[colnames(rna.so),]))
fwrite(x = data_to_write_out, file = "/home/xcx/results/1469/6-Harmony/z_rna.csv")

data_to_write_out <- as.data.frame(as.matrix(harmony_embeddings[colnames(atac.so),]))
fwrite(x = data_to_write_out, file = "/home/xcx/results/1469/6-Harmony/z_atac.csv")
```

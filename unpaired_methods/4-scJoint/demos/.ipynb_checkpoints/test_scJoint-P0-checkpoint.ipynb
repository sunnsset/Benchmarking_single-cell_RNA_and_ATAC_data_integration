```python
label_to_ids = pd.read_csv("/home/xcx/MYBenchmark-codes/4-scJoint/label_to_idx-P0.txt", sep = " ", header = None)

index = {}
for i in range(label_to_ids.shape[0]):
    index[label_to_ids[0][i]] = label_to_ids[1][i]

lists = []

for line in open("/home/xcx/MYBenchmark-datas/P0/cell_label.csv"):
    line=line
    line = line.split(",")[-1].strip()
    # print(line)
    if line == 'cell_type':
        continue
    lists.append(index[line])

with open("/home/xcx/MYBenchmark-codes/4-scJoint/anno-num-P0.txt",'a') as f:
    for i in lists:
        # print(i)
        f.write(str(i)+'\n')
```

```python
import pandas as pd
import scanpy as sc

df = sc.read("/home/xcx/MYBenchmark-datas/P0/ATAC/region2gene.mtx", cache = True)
df = df.X
df = df.todense()
print(df.shape)

atac = sc.read("/home/xcx/MYBenchmark-datas/P0/ATAC/matrix.mtx", cache = True).T
atac = atac.X
atac = atac.todense()
print(atac.shape)

new_atac=np.matmul(atac,df)
print(new_atac.shape)
```

```python
# binary
atac_raw = new_atac
rna_raw = sc.read("/home/xcx/MYBenchmark-datas/P0/RNA/matrix.mtx", cache = True).T
rna_raw = rna_raw.X
rna_raw = rna_raw.todense()

counts_atac = np.where(atac_raw > 0, 1, 0)
counts_rna = np.where(rna_raw > 0, 1, 0)

pd.DataFrame(counts_atac).to_csv("/home/xcx/MYBenchmark-codes/4-scJoint/data/P0/binary_gene_activity_atac.csv", index=None)
pd.DataFrame(counts_rna).to_csv("/home/xcx/MYBenchmark-codes/4-scJoint/data/P0/binary_counts_rna.csv", index=None)
```

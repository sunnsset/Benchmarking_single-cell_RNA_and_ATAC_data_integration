```python
from cobolt.utils import SingleData, MultiomicDataset
from cobolt.model import Cobolt
import os
```

## Read Data

```python
# unpaired RNA
from scipy import io
import pandas as pd
import numpy as np
import scipy
import scanpy as sc

# path = os.path.join(dat_dir)
counts_rna = sc.read("/data1/xcx/unpaired-data/uterus/RNA/highvar_counts_rna.mtx", cache = True).X.toarray()

hihgvar_geneids = pd.read_csv('/data1/xcx/unpaired-data/uterus/highvar_geneids.csv', sep='\t', header=0).values[:,0]
barcode = sc.read("/data1/xcx/unpaired-data/uterus/RNA/Uterus_Wang_2020_selected.h5ad", cache = True).obs_names.values
feature = sc.read("/data1/xcx/unpaired-data/uterus/RNA/Uterus_Wang_2020_selected.h5ad", cache = True).var_names.values
feature = feature[hihgvar_geneids]

unpaired_rna = SingleData("GeneExpr", "scRNA-seq", feature, counts_rna, barcode)
```

```python
# unpaired ATAC
from scipy import io
import pandas as pd
import numpy as np
import scipy
import scanpy as sc

# path = os.path.join(dat_dir)
counts_atac = sc.read("/data1/xcx/unpaired-data/uterus/ATAC/uterus_Zhang2021/highvar_counts_atac.mtx", cache = True)
sc.pp.log1p(counts_atac)
counts_atac = np.array(counts_atac.X.todense())

# count = scipy.sparse.csr_matrix(count)
hihgvar_regionids = pd.read_csv("/data1/xcx/unpaired-data/uterus/highvar_regionids.csv", sep='\t', header=0).values[:,0]
barcode = sc.read("/data1/xcx/unpaired-data/uterus/ATAC/uterus_Zhang2021/adata_anno_selected.h5ad", cache = True).obs_names.values
feature = sc.read("/data1/xcx/unpaired-data/uterus/ATAC/uterus_Zhang2021/adata_anno_selected.h5ad", cache = True).var_names.values
feature = feature[hihgvar_regionids]  

unpaired_atac = SingleData("ChromAccess", "scATAC-seq", feature, np.array(counts_atac), barcode)
```

```python
# paired RNA
from scipy import io
import pandas as pd
import numpy as np
import scipy
import scanpy as sc

dat_dir = "/home/xcx/SNARE-GSE126074/P0/"
path = os.path.join(dat_dir)
counts_rna = sc.read(os.path.join(path, "RNA/matrix.mtx"), cache = True).T
sc.pp.log1p(counts_rna)
counts_rna = np.array(counts_rna.X.todense())

# count = scipy.sparse.csr_matrix(count)
barcode = pd.read_csv(os.path.join(path, "RNA/barcodes.tsv"), header=None, usecols=None).values.astype('str')[:,0]
feature = pd.read_csv(os.path.join(path, "RNA/features.tsv"), header=None, usecols = None).values.astype('str')[:,0]

paired_rna = SingleData("GeneExpr", "SNARE-seq", feature, counts_rna, barcode)
```

```python
# paired ATAC
from scipy import io
import pandas as pd
import numpy as np
import scipy
import scanpy as sc

path = os.path.join(dat_dir)
counts_atac = sc.read(os.path.join(path, "ATAC/matrix.mtx"), cache = True).T
sc.pp.log1p(counts_atac)
counts_atac = np.array(counts_atac.X.todense())

# count = scipy.sparse.csr_matrix(count)
barcode = pd.read_csv(os.path.join(path, "ATAC/barcodes.tsv"), header=None, usecols=None).values.astype('str')[:,0]
feature = pd.read_csv(os.path.join(path, "ATAC/peaks.tsv"), header=None, usecols = None).values.astype('str')[:,0]
                      
paired_atac = SingleData("ChromAccess", "SNARE-seq", feature, np.array(counts_atac), barcode)
```

```python
# Quality filtering on features.
paired_rna.filter_features(upper_quantile=0.99, lower_quantile=0.7)
paired_atac.filter_features(upper_quantile=0.99, lower_quantile=0.7)
```

```python
paired_rna.count = scipy.sparse.csr_matrix(paired_rna.count)
paired_atac.count = scipy.sparse.csr_matrix(paired_atac.count)
unpaired_rna.count = scipy.sparse.csr_matrix(unpaired_rna.count)
unpaired_atac.count = scipy.sparse.csr_matrix(unpaired_atac.count)
```

```python
# multi_dt = MultiomicDataset.from_singledata(
#     mop_mrna, mop_atac, snare_atac, snare_mrna)
multi_dt = MultiomicDataset.from_singledata(unpaired_rna, unpaired_atac, paired_rna, paired_atac)
print(multi_dt)
```

model = Cobolt(dataset=multi_dt, lr=0.005, n_latent=10)
model.train(num_epochs=20)## Training

We now train the Cobolt model with the number of latent variables set to 10. We
use a small iteration in this vignette for illustration purpose. The default
number of iteration is set to 100.

```python
import datetime
starttime = datetime.datetime.now()

model = Cobolt(dataset=multi_dt, lr=0.0001, n_latent=30)
model.train(num_epochs=100)

#long running
endtime = datetime.datetime.now()
print((endtime-starttime).seconds)
```

Calculate the corrected latent variables.

```python
model.calc_all_latent()
```

The latent variables can be accessed through function `get_all_latent`.

```python
latent = model.get_all_latent()
```

The latent variables pre-correction can be accessed by setting the
`correction` parameter. The returned results contain all available posterior
means for each cell. For example, a SNARE-seq cell can have 3 posterior means,
corresponding to the posterior distribution given gene expression, given
chromatin accessibility, and given both modalities.

```python
latent_raw = model.get_all_latent(correction=False)
```

## Clustering and Visualization

We can cluster cells on the latent variables using Leiden algorithm. Higher
resolution gives more clusters, and lower resolution gives fewer number of
clusters.

```python
model.clustering(algo="leiden", resolution=1)
```

Users are encouraged to try different resolution parameters. Calls to
`clustering` function will be stored in the object. Cluster results obtained
with a specific algorithm and resolution and can be accessed as the following.

```python
clusters = model.get_clusters(algo="leiden", resolution=1)
```

Visualize the clustering results on UMAP.

```python
model.scatter_plot(reduc="UMAP", algo="leiden", resolution=1, s=0.2)
```

```python
dir = "/home/xcx/results/1469/4-scJoint/"
import process_db
import h5py
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import random
random.seed(1)

rna_embeddings = np.loadtxt(dir+'binary_counts_rna_embeddings.txt')
atac_embeddings = np.loadtxt(dir+'binary_gene_activity_atac_embeddings.txt')
print(rna_embeddings.shape)
print(atac_embeddings.shape)
embeddings =  np.concatenate((rna_embeddings, atac_embeddings))
print(embeddings.shape)
tsne_results = TSNE(perplexity=30, n_iter = 1000).fit_transform(embeddings)
tsne_results.shape
df = pd.DataFrame()
df['tSNE1'] = tsne_results[:,0]
df['tSNE2'] = tsne_results[:,1]
```

```python
rna_labels = np.loadtxt('/home/xcx/MYBenchmark-codes/4-scJoint/anno-num-1469.txt')
atac_labels = np.loadtxt('/home/xcx/MYBenchmark-codes/4-scJoint/anno-num-1469.txt')
atac_predictions = np.loadtxt(dir+'binary_gene_activity_atac_knn_predictions.txt')
labels = np.concatenate((rna_labels, atac_predictions))
label_to_idx = pd.read_csv('/home/xcx/MYBenchmark-codes/4-scJoint/label_to_idx-1469.txt', sep = '\t', header = None)
label_to_idx.shape
label_dic = []
for i in range(label_to_idx.shape[0]):
    label_dic = np.append(label_dic, label_to_idx[0][i][:-2])
```

```python
common_label_cnt = 0
correct = 0
for i in range(len(atac_labels)):
    if atac_labels[i] >= 0:
        common_label_cnt += 1
        if atac_labels[i] == atac_predictions[i]:
            correct += 1
correct/common_label_cnt
```

```python
data_label = np.array(["RNA-seq", "ATAC-seq"])
df['data'] = np.repeat(data_label, [rna_embeddings.shape[0], atac_embeddings.shape[0]], axis=0)
df['predicted'] = label_dic[labels.astype(int)]
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(
    x = "tSNE1", y = "tSNE2",
    hue = "data",
    palette = sns.color_palette("tab10", 2),
    data = df,
    legend = "full",
    alpha = 0.3
)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(
    x = "tSNE1", y = "tSNE2",
    hue = "predicted",
    palette = sns.color_palette("Set2", 7),
    data = df,
    legend = "full",
    alpha = 0.3
)
```
